```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data=train_test_split(redwine,train_size=0.8)
```

```python
len(train_data)
```

```python
plt.pie(train_data.quality.value_counts(),labels=train_data.quality.value_counts().index)
plt.show()
```

```python
plt.pie(test_data.quality.value_counts(),labels=test_data.quality.value_counts().index)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=len(redwine.columns.tolist()),ncols=2,figsize=(10,3*len(cols)))
for i,col in enumerate(cols):
    current_ax_train = ax[i][0]
    current_ax_test = ax[i][1]

    sns.distplot(train_data[col],ax=current_ax_train)
    sns.distplot(test_data[col],ax=current_ax_test)

fig.tight_layout()
plt.show()
```

```python
def train_test_finder(i):
    '''
    returns 0 if the index is in the test data, and 1 if it is in the train data
    i : int value of an index that is part of the redwine dataframe
    '''
    return int (i in train_data.index.tolist())
```

```python
train_test_finder(138)
```

```python
redwine.loc[:,'train']=pd.Series(redwine.index.tolist()).apply(train_test_finder)
```

```python
feature_cols= ['fixed acidity',
 'volatile acidity',
 'citric acid',
 'residual sugar',
 'chlorides',
 'free sulfur dioxide',
 'total sulfur dioxide',
 'density',
 'pH',
 'sulphates',
 'alcohol',
 ]
```

```python
target_col='quality'
```

```python
X_train = train_data[feature_cols].values
Y_train = train_data[target_col].values
```

```python
X_test = test_data[feature_cols].values
Y_test = test_data[target_col].values
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_test_predicted = model.predict(X_test)
```

```python
Y_test_predicted == Y_test
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
print(classification_report(Y_test,Y_test_predicted))
```

```python
cm = pd.DataFrame(confusion_matrix(Y_test,Y_test_predicted,labels=[3,4,5,6,7,8]),columns=[3,4,5,6,7,8],index=[3,4,5,6,7,8])
cm
```

```python
cm['total']=cm.sum(axis=1)
```

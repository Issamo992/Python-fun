```python
plt.figure(figsize=(15,10))
plt.plot(df.Price.loc["2010":"2011"],".",label="Daily")
plt.plot(df.Price.loc["2010":"2011"].resample("M").mean(), label= "Monthly mean")
plt.plot(df.Price.loc["2010":"2011"].resample("M").max(), label="Monthly max")
plt.plot(df.Price.loc["2010":"2011"].resample("M").min(), label="Monthly min")
plt.legend()
plt.show()
```
